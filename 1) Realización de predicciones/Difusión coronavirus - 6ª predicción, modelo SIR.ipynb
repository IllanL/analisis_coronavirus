# DIFUSIÓN DEL CORONAVIRUS EN ESPAÑA:

## (6ª predicción) modelo SIR (ajuste de ecuaciones desacopladas)

Partiendo de la solución encontrara para el modelo SI, vamos a aprovechar esta resolución para ahora a ajustar las ecuaciones del modelo SIR para realizar predicciones que se ajusten mejor a los datos.

Las ecuaciones del modelo SIR son las siguientes:

$ dS/dt = -\beta \frac{IS}{N} $

$ dI/dt =  \beta \frac{IS}{N} - \gamma I $

$ dR/dt = \gamma I$

Con:

$ dS/dt + dI/dt + dR/dt = 0 $

$ S + I + R = cte. $

El planteamiento consistirá en ajustar $\gamma$ según la curva de recuperados diarios, por un lado:

$ dR/dt = \gamma I$ --> $\gamma = R'/I $ --> $\gamma = \frac {R} {\int_{0}^{t} I dt} $

Y, por otro lado, emplearemos las ecuaciones diferenciales del modelo SI, empleadas en el notebook anterior:

$ I = \frac {1}{A e^{ - \beta x} + 1/N} $

e:

$ I' = \frac {A \beta e^{ - \beta x}}{(A e^{ - \beta x} + 1/N)^2} $

Donde, en este caso, I no representa ya los contagios totales sino solamente los nuevos infectados.

```python
def error_mi_func(variab, x, y, weights):
    
    B, A0, alpha0, A1, alpha1 = variab
    
    y_est = B + A0 * x ** alpha0 + A1 * x **alpha1
    
    return weights * (y - y_est)
```

```python
def error_mi_func_2(variab, x, y, weights):
    
    A0, beta0, K = variab
    
    y_est = 1/(A0 * np.exp(-beta0 * x) + K)
    
    return weights * (y - y_est)
```

## 2.3) MUERTOS

Modificamos ahora, respecto al modelo SI, la forma en la que realizamos las predicciones de la mortalidad.

Siendo estrictos, esta no debería modelarse con la misma ecuación que rige el comportamiento de $I'$ (como hicimos en el notebook del modelo SI), sino que idealmente tendremos que:

$ R = A + M $

Suponiendo una tasa de mortalidad fija tendremos que:

$ M' = k A' $

Integrando:

$ M = k A + C $

Sin embargo, haciendo esta asunción estamos también suponiendo que el tiempo que tarda una persona en fallecer por la enfermedad es el mismo que el que tarda en recuperarse.

Lo que sucede que en la práctica es que los que mueren tienen un tiempo de enfermedad más breve que aquellos a los que finalmente les dan el alta, de forma que la última ecuación será más bien:

$ M (t) = k A (t-\Delta t) + c $

Y, por lo tanto, la tasa de mortalidad será:

$mortalidad = \frac{k}{k+1}$

Vamos a intentar ajustar esta relación:

```python
def error_mortalidad(variab, x, y, weights):
    
    tasa_mort = variab[0]
    Const = variab[1]  
    C = 4
    
    aaa = x.shift(-C)
    pend = x.diff()[-C]
    
    # al aplicar el diff, tenemos que rellenar los valores perdidos con una estimación. Realizamos una estimación lineal 
    # a partir del último valor, y con la última diferencia, para obtener los NaN del diff:

    while aaa.isnull().values.any():
        aaa[-C] = aaa[-C - 1] + pend
        C = C -1

    y_est = tasa_mort * aaa + Const
         
    return weights * (y - y_est)
```

```python
weights = np.ones(datos_full_index.shape[0])

vas = [0, 0]

for i in range(1,5):
    sal = leastsq(error_mortalidad, vas, args=(datos_full_index['ALTAS'], datos_full_index['MUERTOS'], weights))
    
    if sal[1] in range (1,5):
        break
    else:
        vas = sal[0]

coefs_i_2= sal[0]

print(coefs_i_2)

C = 4
aaa = datos_full_index['ALTAS'].shift(-C)
pend = datos_full_index['ALTAS'].diff()[-C]

while aaa.isnull().values.any():

    aaa[-C] = aaa[-C - 1] + pend

    C = C -1

datos_full_index['PRED_MUERTOS'] = coefs_i_2[0] * aaa + coefs_i_2[1]
datos_full_index[['MUERTOS', 'PRED_MUERTOS']].plot()
plt.title('AJUSTE DE LOS MUERTOS');
```

```python
mortalidad = coefs_i_2[0]/(coefs_i_2[0] + 1)
mortalidad
```

Como vemos, tenemos un muy buen ajuste.

El inconveniente de seguir este procedimiento es que tendremos que estimar las altas (hemos introducido un método lineal para predecir dichas altas en la función que optimizamos, al tener que irnos los días del desfase hacia el futuro para traernos el valor presente) para poder realizar proyecciones a futuro de las defunciones.

Por lo tanto, será conveniente buscar otra vía para estimar las mismas.

Lo que en cambio nos ha ofrecido esta aproximación es un dato de la mortalidad real de la enfermedad, (referida a los casos detectados, no al total de infectados, que es desconcido), que como cabía esperar, es más elevada que los datos oficiales, en los que se dividen las defunciones entre el total de contagiados conocidos, cuando de la mayoría de ellos aún no se conoce la evolución de su enfermedad.

**De la relación encontrada sin embargo, que existe un desfase entre los días que una persona pasa ingresada cuando fallece, a los días que pasa ingresada y se le da el alta. O, dicho de otra forma: los fallecimientos de un día serán proporcionales, no al total de ingresados existente, sino al total de ingresados que llevan menos de n días, dado que los que llevan n+1 días y más, son realmente infectados en los tramos finales de recuperación, que ya se encuentran fuera de peligro.**

**Este será nuestro planteamiento para predecir el número de fallecidos: tomaremos la suma de contagios de los últimos cuatro días para predecir los fallecidos del siguiente:**

```python
C_ultimos_n_dias = C_df['C_DIA'].rolling(4).sum()[fecha_inicio:fecha_fin]
C_ultimos_n_dias.plot()
```

```python
def error_mortalidad_2 (variab, x, y, weights):
    
    tasa_mort = variab[0]
    Const = variab[1]

    y_est = tasa_mort * x + Const

            
    return weights * (y - y_est)
```

```python
weights = np.ones(datos_full_index.shape[0])

vas = [0,0]

for i in range(1,5):
    sal = leastsq(error_mortalidad_2, vas, args=(C_ultimos_n_dias, datos_full_index['M_DIA'], weights))
    
    if sal[1] in range (1,5):
        break
    else:
        vas = sal[0]

coefs_i_2= sal[0]

print(coefs_i_2)

datos_full_index['PRED_M_DIA'] = coefs_i_2[0] * C_ultimos_n_dias+ coefs_i_2[1]

datos_full_index[['M_DIA', 'PRED_M_DIA']].plot()
plt.title('AJUSTE DE LOS MUERTOS');
```

Vemos que aparentemente, empleando datos de los infectados durante los últimos 15 días, el ajuste es bueno.

Para contrastar resultados, vamos a ver el ajustar la ecuación que predice los infectados diarios, a los muertos diarios, y comparar los resultados con los recién obtenidos:

```python
def error_mortalidad_3(variab, x, y, weights):
    
    A0, beta0, K = variab
    
    y_est = A0 * beta0 * np.exp(-beta0 * x)/(A0 * np.exp(-beta0 * x) + K)**2
    
    return weights * (y - y_est)
```

```python
weights = np.ones(datos_full_index.shape[0])

vas = [0.0005, 0.0005, 0.0005]

for i in range(1,5):
    sal = leastsq(error_mortalidad_3, vas, args=(datos_full_index['DIA'], datos_full_index['M_DIA'], weights))
    
    if sal[1] in range (1,5):
        break
    else:
        vas = sal[0]

coefs_i_3 = sal[0]

print(coefs_i_3)

# datos_full_index['PRED_MUERTOS'] = coefs_i_2[0] * datos_full_index['R'] + coefs_i_2[1]


A0, beta0, K = coefs_i_3
x = datos_full_index['DIA']
datos_full_index['PRED_M_DIA_2'] = A0 * beta0 * np.exp(-beta0 * x)/(A0 * np.exp(-beta0 * x) + K)**2

datos_full_index[['M_DIA', 'PRED_M_DIA', 'PRED_M_DIA_2']].plot()
plt.title('AJUSTE DE LOS MUERTOS');
```

Vemos que ambas curvas se parecen, pero que se obtiene un mejor ajuste con la idea incial, que es la que conservaremos.
